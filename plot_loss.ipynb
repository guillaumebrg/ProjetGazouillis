```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# file_path = "results/exp_002/log.txt"
file_pathes = ['results/exp_001/log.txt','results/exp_002/log.txt','results/exp_003/log.txt','results/exp_004/log.txt',]

plt.figure()
for file_path in file_pathes:
    times = []
    losses = []

    f = open(file_path,'r')
    for line in f:
        arr = line.split()
#         Get bacth size
        if arr[0]=="Batch":
            batch_size = int(arr[3])
#         Get val loss and time of epoch
        if arr[0]=="Epoch":
            losses.append(float(arr[11]))
            times.append(int(arr[3]))

    plt.plot(np.cumsum(np.array(times)),losses,label='Batch size {}'.format(batch_size))
    print "minimum loss : ",np.min(losses)
    f.close()
# plt.xlim(0,30000)
plt.ylim(1.6,2.2)
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("val loss")
```
