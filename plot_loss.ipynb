```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = "results/exp_002/log.txt"

times = []
losses = []

f = open(file_path,'r')
for line in f:
    arr = line.split()
    if arr[0]=="Epoch":
        losses.append(arr[11])
        times.append(int(arr[3]))
        
plt.figure()
plt.plot(np.cumsum(np.array(times)),losses)
```
