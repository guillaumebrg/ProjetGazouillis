```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# file_path = "results/exp_002/log.txt"
file_pathes = ['results/exp_006/log.txt','results/exp_009/log.txt','results/exp_010/log.txt','results/exp_011/log.txt']

plt.figure()
for file_path in file_pathes:
    times = []
    losses = []

    f = open(file_path,'r')
    for line in f:
        arr = line.split()
#         Get bacth size
        if arr[0]=="LR":
            lr = float(arr[2])
#         Get val loss and time of epoch
        if arr[0]=="Epoch":
            if not np.isnan(float(arr[11])):
                losses.append(float(arr[11]))
                times.append(int(arr[3]))

    plt.plot(np.cumsum(np.array(times)),losses,label='Learning rate = {}'.format(lr))
    print "minimum loss : ",np.min(losses)
    f.close()
#plt.xlim(0,40000)
plt.ylim(1.6,4)
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("val loss")
```
